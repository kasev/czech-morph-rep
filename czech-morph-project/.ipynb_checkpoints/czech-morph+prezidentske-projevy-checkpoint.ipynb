```python
import sys
import nltk
import pandas as pd
from ufal.morphodita import *
```

```python
czech_tagger='czech-morfflex-pdt-161115.tagger'
tagger = Tagger.load(czech_tagger)
```

```python
forms = Forms()
lemmas = TaggedLemmas()
tokens = TokenRanges()
tokenizer = tagger.newTokenizer()
```

```python
### příklad

line = "Viděl jsem příliš žluťoučkého koně pět ďábelské ódy na listy lip. Celkem se mi to líbilo, ale víc nedokážu říct"
```

```python
string_variable = "Tento text je nyní obsahem proměnné. Mohu libovolně přidávat slova. Může být libovolně dlouhý, obsahovat mnoho vět, slovo za slovem, dokud jej neukončí uvozovky. V naší ukázce, budeme počítat termín slovo. Kolikrát se zde termín slovo objevuje?"
```

```python
print(line)
```

```python
tokenizer.setText(string_variable)
t = 0
output = []

while tokenizer.nextSentence(forms, tokens):
      tagger.tag(forms,lemmas)
      
      for i in range (len(lemmas)):
        form = forms[i]
        output.append([form, lemmas[i].lemma, lemmas[i].tag])
```

```python
output_df = pd.DataFrame(output, columns=("form", "lemma", "TAG"))
output_df
```

```python
line_lemmata = output_df["lemma"].tolist()
```
