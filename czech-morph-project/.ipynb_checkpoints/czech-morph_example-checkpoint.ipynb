```python
import sys
import nltk
import pandas as pd
from ufal.morphodita import *
```

```python
czech_tagger='czech-morfflex-pdt-161115.tagger'
tagger = Tagger.load(czech_tagger)
```

```python
forms = Forms()
lemmas = TaggedLemmas()
tokens = TokenRanges()
tokenizer = tagger.newTokenizer()
```

```python
### příklad

line = "Viděl jsem příliš žluťoučkého koně pět ďábelské ódy na listy lip. Celkem se mi to líbilo, ale víc nedokážu říct"
```

```python
tokenizer.setText(line)
t = 0
output = []

while tokenizer.nextSentence(forms, tokens):
      tagger.tag(forms,lemmas)
      
      for i in range (len(lemmas)):
        form = forms[i]
        output.append([form, lemmas[i].lemma, lemmas[i].tag])
```

```python
output_df = pd.DataFrame(output, columns=("form", "lemma", "TAG"))
output_df
```
